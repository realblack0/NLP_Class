### 음절 함수

```python
def umjeol(sentence, n=2):
    ngram = []
    tokens = list(sentence.replace(" ", "_"))
    for i in range(len(tokens)-(n-1)):
        ngram.append("".join(tokens[i:i+n]))
    return ngram
```

### ngram 함수

```python
from collections import defaultdict

def ngram(sentence):
    gram = defaultdict(int) # defaultdict
    
    # spjlit 대신에 re.findall(r"\b\w\b\s") 을 써도 된다. 정규식 방법.
    for _ in sentence.split():
        for n in range(1, len(_)+1):
            for token in umjeol(_, n):
                gram[token] += 1
    return gram
```

```python
ngram("안녕하세요")
```

```python
from konlpy.corpus import kolaw
from nltk.tokenize import sent_tokenize

corpus = kolaw.open(kolaw.fileids()[0]).read()
tokens = defaultdict(int)
for _ in sent_tokenize(corpus):
    for k,v in ngram(_).items():
        tokens[k] += v
```

# n이 1일때 n-gram

```python
# unigram
# {uni-token:경우의 수} 쌍이다.

unigram = {k:v for k,v in tokens.items() if len(k) == 1}
unigram
```

```python
# uni-token 총 갯수

sum_unigram = sum(unigram.values())
sum_unigram
```

```python
# {uni-token:발생확률}쌍이다. 
# 경우의 수를 전체 경우의 수로 나누면 확률이 된다.

p_unigram = {k:v/sum_unigram for k,v in tokens.items()
            if len(k) == 1}
p_unigram
```

```python
# 확률이 높은순 상위 5개.

sorted(p_unigram.items(), key=lambda x:x[1], reverse=True)[:5]
```

unigram(1-gram)
    - 문제점1: 조건부 확률을 알 수 없다.
        - 조건부 확률: P(A) P(B) P(C) P(D) = P(A,B,C | D)
    - 문제점2: 언어는 앞에 어떤 단어가 나왔는지 순서가 중요한데, unigram은 순서를 바꿔도 확률값이 동일하다.
        - P(A) * P(B) = P(B) * P(A)
    
bigram(2-gram)
    - unigram의 문제점을 해결하기 위해서 바로 앞의 단어 1개까지만 순서를 생각해보자는 개념.

# n-gram을 이용한 단어 예측

```python
# 수업시간에 짠 코드

def findNext(t, n):
    given = tokens[t]
    return {k:v/given for k, v in tokens.items()
            if len(k) == n and k.startswith(t)}
```

```python
# 확률의 총 합은 1이 되어야한다.

sum(findNext("국회", 4).values())
```

> 확률의 총 합은 1이 되어야 하는데 1이 안 나온다.
>
> 현재 given은 t의 경우의 수이다. 확률로 나타내기 위해서는 given을 **t로 시작하고 길이가 n인 token**의 총 갯수로 바꿔야한다. 

```python
# 수정한 함수
# given은 주어진 조건에 대한 총 경우의 수 이어야한다.

def findNext2(t, n):
    given = sum([v for k, v in tokens.items() 
                 if k.startswith(t) and len(k) == n])
    return {k:v/given for k, v in tokens.items()
            if len(k) == n and k.startswith(t)}
```

```python
# 확률의 총 합은 1이 되어야한다.

sum(findNext2("국회", 4).values())
```

```python
# 메모리 효율을 위해서 한번 더 수정한 함수

def findNext3(t, n):
    given = {k:v for k, v in tokens.items() 
                 if k.startswith(t) and len(k) == n}
    return {k:v/sum(given.values()) for k, v in given.items()}
```

```python
%%timeit

findNext2("국회", 4).values()
```

```python
%%timeit

findNext3("국회", 4).values()
```

> findNext2에 비해 findNext3이 거의 2배 빠르다.

```python
# "국"으로 시작하고, 2음절로 구성된 token 중에, 확률 상위 5개 나열

sorted(findNext3("국", 2).items(), key=lambda x:x[1], reverse=True)[:5]
```

```python
# "국회"로 시작하고, 3음절로 구성된 token 중에, 확률 상위 5개 나열

sorted(findNext3("국회", 3).items(), key=lambda x:x[1], reverse=True)[:5]
```

```python
# "국회"로 시작하고, 4음절로 구성된 token 중에, 확률 상위 5개 나열

sorted(findNext3("국회", 4).items(), key=lambda x:x[1], reverse=True)[:5]
```
