unigram(1-gram)
    - P(A) P(B) P(C) P(D) = P(A,B,C | D)
    - 문제점: 언어는 앞에 어떤 단어가 나왔는지 순서가 중요한데, unigram은 순서를 바꿔도 확률값이 동일하다.
        - P(A) * P(B) = P(B) * P(A)
    
bigram(2-gram)
    - unigram의 문제점을 해결하기 위해서 바로 앞의 단어 1개까지만 순서를 생각해보자는 개념.

unigram + bigram
    - 섞어 씀으로써 문제를 해결한다.... (무슨문제?)

```python
def umjeol(sentence, n=2):
    ngram = []
    tokens = list(sentence.replace(" ", "_"))
    for i in range(len(tokens)-(n-1)):
        ngram.append("".join(tokens[i:i+n]))
    return ngram
```

```python
from collections import defaultdict

def ngram(sentence):
    gram = defaultdict(int) # defaultdict
    
    # spjlit 대신에 re.findall(r"\b\w\b\s") 을 써도 된다. 정규식 방법.
    for _ in sentence.split():
        for n in range(len(_)):
            for token in umjeol(_, n):
                gram[token] += 1
#                 print(token)
#                 print(gram)
    return gram
```

```python
from konlpy.corpus import kolaw
from nltk.tokenize import sent_tokenize
corpus = kolaw.open(kolaw.fileids()[0]).read()
tokens = defaultdict(int)
for _ in sent_tokenize(corpus):
    for k,v in ngram(_).items():
        tokens[k] += v
```

```python
# 수정

def findNext(t, n):
    given = tokens[t] # 바뀐 부분
    return {k:v/given for k, v in tokens.items()
            if len(k) == n and k.startswith(t)}
```

```python
# "국"으로 시작하고, 2음절로 구성된 token 중에, 상위 5개 나열

sorted(findNext("국", 2).items(), key=lambda x:x[1], reverse=True)[:5]
```
